# Recommendations + Content Calendar for Alfido Tech

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
from pathlib import Path

data_path = Path('../data/exports')

master_df = pd.read_csv(data_path / 'master_df.csv')
```

```python
master_df.head()
```

```python
time_df = master_df.copy()
time_df['created_time'] = pd.to_datetime(time_df['created_time'], errors='coerce')
time_df['hour'] = time_df['created_time'].dt.hour
time_df['total_engagement'] = time_df['total_likes'] + time_df['total_comments']
time_df['day_of_week_num'] = time_df['created_time'].dt.dayofweek

print("Recommendations for Alfido Tech ")
# =======================================================
print(f"Time range : {time_df['created_time'].min()} to {time_df['created_time'].max()}")
print("Total Samples :\n")

#==================== DESCRIBING THE DATA ===========================
display(time_df[['created_time', 'hour', 'total_engagement']].describe())
print("\nSample:")


#====================== DISPLAYING THE DATA ============================
display(time_df[['created_time', 'hour', 'day_of_week_num', 'total_engagement']].head(5))

best_hour = time_df.groupby('hour')['total_engagement'].mean().idxmax()
print(f"Best hour for engagement: {best_hour}")

best_day = time_df.groupby('day_of_week_num')['total_engagement'].mean().idxmax()
print(f"Best day of the week for engagement: {best_day}")   

print(f"Total Engagement by Hour:\n{time_df.groupby('hour')['total_engagement'].sum()}")
```
