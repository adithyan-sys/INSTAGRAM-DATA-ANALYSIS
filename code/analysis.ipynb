```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
data_path = '../data/'
photos = pd.read_csv(data_path + 'photos.csv')
comments = pd.read_csv(data_path + 'comments.csv')
likes = pd.read_csv(data_path + 'likes.csv')
follows = pd.read_csv(data_path + 'follows.csv')
photo_tags = pd.read_csv(data_path + 'photo_tags.csv')
tags = pd.read_csv(data_path + 'tags.csv')
users = pd.read_csv(data_path + 'users.csv')

def normalize_columns(df):
    df = df.copy()
    df.columns = df.columns.str.strip().str.lower().str.replace(r'\s+', '_', regex=True)
    return df

photos = normalize_columns(photos)
comments = normalize_columns(comments)
likes = normalize_columns(likes)
follows = normalize_columns(follows)
photo_tags = normalize_columns(photo_tags)
tags = normalize_columns(tags)
users = normalize_columns(users)
print("All files are loaded successfully!!")
```

```python
print("Photos shape: ", photos.shape)
print("Comments shape: ", comments.shape)
print("Likes shape: ", likes.shape)
print("Follows shape: ", follows.shape)
print("Photo Tags shape: ", photo_tags.shape)
print("Tags shape: ", tags.shape)
print("Users shape: ", users.shape)
```

```python
tables = {
    'photos': photos,
    'comments': comments,
    'likes': likes,
    'follows': follows,
    'photo_tags': photo_tags,
    'tags': tags,
    'users': users
}
# for name, df in tables.items():
#     print(f"\n{name.upper()}--> shape: {df.shape}")
#     print(df.columns.tolist())
#     print(df.head(2))

print("📊 INSTAGRAM DATASET SUMMARY\n" + "="*60)

for name, df in tables.items():
    print(f"\n🔹 {name.upper()} TABLE")
    print("-" * 50)
    print(f"Shape (Rows x Columns) : {df.shape[0]:,} rows × {df.shape[1]} columns")
    print(f"Columns ({len(df.columns)}):")
    
    # Print columns neatly
    for i, col in enumerate(df.columns, 1):
        print(f"   {i:2d}. {col}")
    
    print("\nFirst 2 Rows:")
    display(df.head(5))   # This shows nice formatted table
    print("-" * 60)
```

```python
# Build a reusable master table for analysis
# The normalized CSV keys are: id, user_id, photo, tag_id, follower, followee
master_df = photos.copy()

# 1) Add user profile details to each post
if 'user_id' in master_df.columns and 'id' in users.columns:
    master_df = master_df.merge(users, left_on='user_id', right_on='id', how='left', suffixes=('', '_user'))
    master_df = master_df.drop(columns=['id_user'], errors='ignore')

# 2) Aggregate hashtags per post
if {'photo', 'tag_id'}.issubset(photo_tags.columns) and {'id', 'tag_text'}.issubset(tags.columns):
    tags_info = photo_tags.merge(tags[['id', 'tag_text']], left_on='tag_id', right_on='id', how='left')
    hashtag_stats = tags_info.groupby('photo').agg(
        hashtag_count=('tag_text', 'count'),
        hashtags=('tag_text', lambda x: ', '.join(sorted(set(x.dropna()))))
    ).reset_index()
    hashtag_stats = hashtag_stats.rename(columns={'photo': 'id'})
    master_df = master_df.merge(hashtag_stats, on='id', how='left')

# 3) Add likes and comments counts
if 'photo' in likes.columns:
    likes_count = likes.groupby('photo').size().reset_index(name='total_likes')
    likes_count = likes_count.rename(columns={'photo': 'id'})
    master_df = master_df.merge(likes_count, on='id', how='left')

if 'photo_id' in comments.columns:
    comments_count = comments.groupby('photo_id').size().reset_index(name='total_comments')
    comments_count = comments_count.rename(columns={'photo_id': 'id'})
    master_df = master_df.merge(comments_count, on='id', how='left')

# 4) Add follower/following metrics from follows for user-level analysis
if {'follower', 'followee'}.issubset(follows.columns):
    followers_received = follows.groupby('followee').size().reset_index(name='followers_received')
    following_count = follows.groupby('follower').size().reset_index(name='following_count')
    master_df = master_df.merge(followers_received, left_on='user_id', right_on='followee', how='left')
    master_df = master_df.merge(following_count, left_on='user_id', right_on='follower', how='left')
    master_df = master_df.drop(columns=['followee', 'follower'], errors='ignore')

# Safe defaults for numeric columns
for col_name in ['hashtag_count', 'total_likes', 'total_comments', 'followers_received', 'following_count']:
    if col_name not in master_df.columns:
        master_df[col_name] = 0
    master_df[col_name] = master_df[col_name].fillna(0).astype(int)

# 5) Create analysis-ready metrics
master_df['total_engagement'] = master_df['total_likes'] + master_df['total_comments']
master_df['engagement_per_follower'] = master_df['total_engagement'] / master_df['followers_received'].replace(0, np.nan)

print('Master table created successfully')
print('Shape:', master_df.shape)
display(master_df.head(257))
```

```python
from pathlib import Path

# Export the analysis-ready master table to CSV
output_dir = Path('../data/exports')
output_dir.mkdir(parents=True, exist_ok=True)

master_csv_path = output_dir / 'master_df.csv'
master_df.to_csv(master_csv_path, index=False)

print(f'Master dataframe saved to: {master_csv_path.resolve()}')
print(f'Rows: {master_df.shape[0]}, Columns: {master_df.shape[1]}')

## Optional source tables too if you want them as separate CSVs  ##
# =======================================================================
# photos.to_csv(output_dir / 'photos_clean.csv', index=False)
# comments.to_csv(output_dir / 'comments_clean.csv', index=False)
# likes.to_csv(output_dir / 'likes_clean.csv', index=False)
# follows.to_csv(output_dir / 'follows_clean.csv', index=False)
# photo_tags.to_csv(output_dir / 'photo_tags_clean.csv', index=False)
# tags.to_csv(output_dir / 'tags_clean.csv', index=False)
# users.to_csv(output_dir / 'users_clean.csv', index=False)
# ===========================================================================


master_df.to_csv(output_dir / 'master_df.csv', index=False)
print('Also saved cleaned source tables into the same exports folder.')
```
